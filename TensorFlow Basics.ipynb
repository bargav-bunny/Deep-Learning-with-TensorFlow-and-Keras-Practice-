```python
import tensorflow as tf
print(tf.__version__)
```

*Initialization of tensors*

```python
x = tf.constant(4)
print(x)
```

```python
x = tf.constant(4.0)
print(x)
```

```python
x = tf.constant(4, shape = (1,1), dtype = tf.float32)
print(x)
```

```python
x = tf.constant([[1,2,3],[4,5,6]])
print(x)
```

```python
x = tf.ones((3,3))
print(x)
```

```python
x = tf.zeros((2,3))
print(x)
```

```python
x = tf.eye(3)
print(x)
```

```python
x = tf.random.normal((3,3), mean=0, stddev=1)
print(x)
```

```python
x = tf.random.uniform((1,3), minval=0, maxval=1)
print(x)
```

```python
x = tf.range(9)
print(x)
```

```python
x = tf.range(start = 1, limit = 10, delta = 2)
print(x)
```

```python
x = tf.cast(x, dtype = tf.float64)
print(x)
```

*Mathematical Operations*

```python
x = tf.constant([1,2,3])
y = tf.constant([9,8,7])
```

```python
z = tf.add(x, y)
print(z)
```

```python
z = x+y
print(z)
```

```python
z = tf.divide(x, y)

z = x/y

print(z)
```

```python
z = tf.multiply(x,y)

z = x*y

print(z)
```

```python
z = tf.tensordot(x, y, axes = 1)

z = tf.reduce_sum(x*y, axis = 0)

print(z)
```

```python
z = x**5

print(z)
```

```python
x = tf.random.normal((2,3))
y = tf.random.normal((3,4))

z = tf.matmul(x,y)

z = x @ y

print(z)
```

*Indexing*

```python
x = tf.constant([0,1,1,2,3,4,1,2])

print(x[1:])
```

```python
print(x[1:3])
```

```python
print(x[:6:2])
```

```python
print(x[::2])
```

```python
print(x[::-2])
```

```python
print(x[::-1])
```

```python
indices = tf.constant([0,3,4])

print(tf.gather(x, indices))
```

*Reshaping*

```python
x = tf.range(9)

x = tf.reshape(x, (3,3))

print(x)
```

```python
x = tf.transpose(x)
print(x)
```
